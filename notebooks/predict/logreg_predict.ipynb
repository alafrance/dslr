# Dataset test : raw + preproccessed

```python
import pandas as pd
import numpy as np
import pickle

dataset = "../../data/raw/dataset_test.csv"
# prend en paramètre le dataset et créer un fichier house.csv avec ce format : Index,Hogwarts House

data = pd.read_csv(dataset)
data
```

```python
def standardize(x):
    means = np.mean(x, axis=0)
    stds = np.std(x, axis=0)

    stds[stds == 0] = 1
    return (x - means) / stds

def impute_missing_values(df):
    return df.apply(lambda col: col.fillna(col.mean()), axis=0)

test_data = standardize(impute_missing_values(data.drop(columns = ['First Name', 'Last Name', 'Birthday', 'Best Hand', 'Index', 'Astronomy', 'Arithmancy', 'Care of Magical Creatures', 'Hogwarts House'])))

test_data
```

# Dataset train (for sklearn) : raw + preproccessed

```python
train_dataset = "../../data/raw/dataset_train.csv"

data = pd.read_csv(train_dataset)
train_data = standardize(impute_missing_values(data.drop(columns = ['Hogwarts House', 'First Name', 'Last Name', 'Birthday', 'Best Hand', 'Index', 'Astronomy', 'Arithmancy', 'Care of Magical Creatures'])))
y_true = {
    'gryffindor': (data['Hogwarts House'] == 'Gryffindor').astype(int),
    'slytherin': (data['Hogwarts House'] == 'Slytherin').astype(int),
    'ravenclaw': (data['Hogwarts House'] == 'Ravenclaw').astype(int),
    'hufflepuff': (data['Hogwarts House'] == 'Hufflepuff').astype(int)
}
```

# Predict

### Load theta train

```python
store_filename_theta = "../../results/models/logistic_regression_model.pkl"

def load_theta_train():
    with open(store_filename_theta, 'rb') as handle:
        result = pickle.load(handle)
    print("Theta_train loaded:", result)
    return result

theta_train = load_theta_train()
```

### Hypothesis
The hypothesis function for logistic regression is defined as follows
With theta represents the weights of the features and x represents the features of the sample.

```python
def hypothesis(x, theta):
    return 1 / (1 + np.exp(-x @ theta))
```

### Predict function

```python
def predict(x, theta):
    bias = np.ones((x.shape[0], 1))
    x_bias = np.hstack([bias, x])
    probabilities = {
        'gryffindor': hypothesis(x_bias, theta['gryffindor']),
        'slytherin': hypothesis(x_bias, theta['slytherin']),
        'ravenclaw': hypothesis(x_bias, theta['ravenclaw']),
        'hufflepuff': hypothesis(x_bias, theta['hufflepuff'])
    }
    return pd.DataFrame(probabilities)

predictions = predict(test_data, theta_train)

predictions
    
```

### Comparaison with Sklearn

```python
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score

sgd = SGDClassifier(loss='log_loss', learning_rate='constant', max_iter=100, tol=1e-7, random_state=42, eta0=0.1)
sgd.fit(train_data, y_true['gryffindor'])

binaray_predictions = accuracy_score(sgd.predict(test_data, theta_train['gryffindor'].apply(lambda x: 1 if x > 0.5 else 0)))
print(accuracy_score(sgd.predict(test_data), binaray_predictions))
```

## Store result in house.csv

```python
store_result_filename = "../../data/trained/house.csv"


predictions_filtered = predictions.idxmax(axis=1).reset_index()
predictions_filtered.columns = ['Index', 'Hogwarts House']
predictions_filtered.to_csv(store_result_filename, index=False, header=['Index', 'Hogwarts House'])
predictions_filtered
```
