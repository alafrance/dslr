# Data 

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('../data/dataset_train.csv')
df
```

# Data histogram
Answer this question : Which Hogwarts course has a homogeneous score distribution between all four houses?

## Group data

```python
group_house = df.groupby('Hogwarts House')
```

## Course

```python
courses = []
for series_name, series in df.items():
    if series.dtype == 'float64':
        courses.append(series_name)
courses
```

## Histogram

```python
import seaborn as sns
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)
plt.figure(figsize=(20,15))
for i, course in enumerate(courses, 1):
    plt.subplot(4, 4, i)
    color_each_house = {'Ravenclaw': 'blue', 'Slytherin': 'green', 'Gryffindor': 'red', 'Hufflepuff': 'yellow'}
    for house in df['Hogwarts House'].unique():
        subset = df[df['Hogwarts House'] == house]
        sns.histplot(subset[course], kde=True, label=house)
    plt.xlabel('Valeurs')
    plt.ylabel('Fréquence')
    plt.title(course)
    plt.legend()
plt.tight_layout()
plt.show()
```
