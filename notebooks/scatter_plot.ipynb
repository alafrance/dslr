## Scatter plot
Question : What are the two features that are similar ?

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

correlation_matrix = df.select_dtypes(include=[np.float64]).corr()
sns.heatmap(correlation_matrix, vmin=-1, vmax=1, annot=True, fmt=".2f", cmap='coolwarm')
plt.show()
```

```python
import numpy as np
corr = df.select_dtypes(include=[np.float64]).corr()
corr.style.background_gradient(cmap='coolwarm')
```

En Ã©tudiant cette matrice de corrÃ©lation, on peut voir que les features `Astronomy` et `Defense Against the Dark Arts` sont trÃ¨s similaires.
Tracons un scatter plot pour confirmer cette hypothÃ¨se.

## Scatter plot

```python
import matplotlib.pyplot as plt
plt.scatter(df['Astronomy'], df['Defense Against the Dark Arts'])
plt.xlabel('Astronomy')
plt.ylabel('Defense Against the Dark Arts')
plt.title('Scatter plot between Astronomy and Defense Against the Dark Arts')
plt.show()
```
