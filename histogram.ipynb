# Data 

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('data/dataset_train.csv')
df
```

# Data histogram
Answer this question : Which Hogwarts course has a homogeneous score distribution between all four houses?

## Group data

```python
group_house = df.groupby('Hogwarts House')
```

## Course

```python
courses = []
for series_name, series in df.items():
    if series.dtype == 'float64':
        courses.append(series_name)
courses
```

## Histogram

```python
plt.figure()
for course in courses:
    plt.figure()
    ax = plt.gca()
    color_each_house = {'Ravenclaw': 'blue', 'Slytherin': 'green', 'Gryffindor': 'red', 'Hufflepuff': 'yellow'}
    for house in group_house.groups:
        ax.hist(group_house.get_group(house)[course], bins=10, color=color_each_house[house], alpha=0.5, label=house)
    plt.xlabel('Valeurs')
    plt.ylabel('FrÃ©quence')
    plt.title(course)
    plt.legend()
    plt.show()
# plt.hist(ravenclaw, bins=10, range=(-500, 500), alpha=0.5, label='Ravenclaw')
# plt.xlabel('Valeurs')
# plt.ylabel('FrÃ©quence')
# plt.title('Ravenclaw')
# plt.show()
```
